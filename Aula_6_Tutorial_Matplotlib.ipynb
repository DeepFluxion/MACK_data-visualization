# Aula Introdutória: Matplotlib para Iniciantes

## 1. O que é Matplotlib?

Matplotlib é uma biblioteca Python para criar visualizações de dados. É como ter uma "caixa de ferramentas" para desenhar gráficos que nos ajudam a entender melhor nossos dados.

## 2. Anatomia de um Gráfico

Todo gráfico matplotlib tem alguns elementos principais:

- **Figure**: A "janela" completa onde o gráfico aparece
- **Axes**: A área onde os dados são plotados (com coordenadas x e y)
- **Axis**: Os eixos x e y com suas marcações
- **Artists**: Todos os elementos visuais (linhas, texto, etc.)

```python
import matplotlib.pyplot as plt
import numpy as np

# Exemplo mais simples possível
fig, ax = plt.subplots()  # Cria figure e axes
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plota os dados
plt.show()  # Mostra o gráfico
```

## 3. Por que usar gráficos?

Os gráficos nos ajudam a:
- **Ver padrões** nos dados
- **Comparar** valores
- **Comunicar** resultados de forma clara
- **Detectar** problemas ou outliers

## Introdução

O Matplotlib é uma das bibliotecas de visualização mais poderosas e versáteis do Python. Este tutorial aborda especificamente os gráficos de dados pareados, onde temos conjuntos de dados (x, y) que representam relacionamentos entre duas variáveis.

### Por que usar gráficos de dados pareados?
- **Visualizar relações**: Entender como uma variável afeta outra
- **Identificar padrões**: Descobrir tendências e correlações nos dados
- **Comparar conjuntos**: Analisar múltiplas séries de dados simultaneamente
- **Comunicar resultados**: Apresentar descobertas de forma clara e impactante

## 3. bar(x, height) - Gráficos de Barras

Os gráficos de barras são perfeitos para comparar valores categóricos.

### Barras Verticais Básicas

```python
# Dados para gráfico de barras
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color='skyblue', alpha=0.7)
plt.title('Gráfico de Barras Básico')
plt.xlabel('Categorias')
plt.ylabel('Valores')

# Adicionar valores no topo das barras
for bar in bars:
```

````python
# Criando um dashboard com todos os tipos de gráfico
fig = plt.figure(figsize=(16, 12))

# Dados para o dashboard
x_dash = np.linspace(0, 10, 50)
y_dash = np.sin(x_dash) + 0.1 * np.random.randn(50)

# 1. Gráfico de linha
plt.subplot(4, 2, 1)
plt.plot(x_dash, y_dash, 'b-', linewidth=2)
plt.title('Plot')
plt.grid(True, alpha=0.3)

# 2. Scatter plot
plt.subplot(4, 2, 2)
plt.scatter(x_dash[::3], y_dash[::3], alpha=0.7, s=50)
plt.title('Scatter')
plt.grid(True, alpha=0.3)

# 3. Bar plot
plt.subplot(4, 2, 3)
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 25, 20]
plt.bar(categories, values, color='lightgreen', alpha=0.7)
plt.title('Bar')

# 4. Stem plot
plt.subplot(4, 2, 4)
x_stem_dash = np.arange(0, 8)
y_stem_dash = np.random.randint(1, 10, 8)
plt.stem(x_stem_dash, y_stem_dash, basefmt=' ')
plt.title('Stem')
plt.grid(True, alpha=0.3)

# 5. Fill between
plt.subplot(4, 2, 5)
y_upper = y_dash + 0.3
y_lower = y_dash - 0.3
plt.plot(x_dash, y_dash, 'b-', linewidth=2)
plt.fill_between(x_dash, y_lower, y_upper, alpha=0.3)
plt.title('Fill Between')
plt.grid(True, alpha=0.3)

# 6. Stackplot
plt.subplot(4, 2, 6)
x_stack_dash = np.arange(1, 8)
y1_dash = np.random.randint(2, 8, 7)
y2_dash = np.random.randint(1, 6, 7)
plt.stackplot(x_stack_dash, y1_dash, y2_dash, alpha=0.7)
plt.title('Stackplot')

# 7. Stairs
plt.subplot(4, 2, 7)
values_dash = [3, 7, 2, 9, 5, 4]
edges_dash = np.arange(len(values_dash) + 1)
plt.stairs(values_dash, edges_dash, fill=True, alpha=0.7)
plt.title('Stairs')
plt.grid(True, alpha=0.3)

# 8. Combinado
plt.subplot(4, 2, 8)
plt.plot(x_dash, y_dash, 'b-', linewidth=2, label='Linha')
plt.scatter(x_dash[::5], y_dash[::5], color='red', s=50, label='Pontos')
plt.fill_between(x_dash, y_dash, alpha=0.3)
plt.title('Combinado')
plt.legend(fontsize=8)
plt.grid(True, alpha=0.3)

plt.suptitle('Dashboard - Todos os Tipos de Gráfico', fontsize=16)
plt.tight_layout()
plt.show()
````
